```python
import cv2
import numpy as np
import time

cap = cv2.VideoCapture(0, cv2.CAP_V4L2)  # Linux
# cap = cv2.VideoCapture(0, cv2.CAP_DSHOW)  # Windows, DirectShow backend, untested

w, h, fps = 320, 240, 10

cap.set(cv2.CAP_PROP_FRAME_WIDTH, w)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, h)
cap.set(cv2.CAP_PROP_FPS, fps)
buffer_size = 10
frame_buffer = np.zeros((buffer_size, h, w), dtype=np.uint8)
idx = 0
```

```python
try:
    while True:
        t_start = time.time()
        ret, frame = cap.read()
        if not ret:
            break

        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        gray = cv2.resize(gray, (w, h))

        frame_buffer[idx % buffer_size] = gray
        idx += 1

        if idx >= buffer_size:
            oldest_idx = idx % buffer_size
            newest_idx = (idx - 1) % buffer_size
            diff = cv2.absdiff(frame_buffer[newest_idx], frame_buffer[oldest_idx])
            cv2.imshow('Difference (oldest vs newest)', diff)

        cv2.imshow('Current', gray)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

        t_elapsed = time.time() - t_start
        if t_elapsed < 1 / fps:
            time.sleep(1 / fps - t_elapsed)
        print("FPS:", cap.get(cv2.CAP_PROP_FPS))
        print("Width:", cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        print("Height:", cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
finally:
    cap.release()
    cv2.destroyAllWindows()
```
